```python
#import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#read img
img = cv2.imread("Professional_Headshot_55.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap='gray')
#resize img
resized_img = cv2.resize(gray_img, (960, 540))
```

```python
#smoothing img
blur_img = cv2.medianBlur(gray_img, 7)

plt.imshow(blur_img, cmap='gray')
#resize img
resized_img = cv2.resize(blur_img, (960, 540))
```

```python
#Retrieving the edges of an image
getedge = cv2.adaptiveThreshold(blur_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
plt.imshow(getedge,cmap='gray')
```

```python
color_img = cv2.bilateralFilter(img, 9, 300, 300)
plt.imshow(color_img,cmap='gray')
```

```python
carton_img = cv2.bitwise_and( color_img ,color_img, mask = getedge,)
plt.imshow(carton_img)
```
